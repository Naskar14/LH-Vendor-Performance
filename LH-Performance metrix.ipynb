```python
lh_leg_cnt = lh_upd['fleet_vendor1'].value_counts()
lh_leg_cnt
```

```python
lh2=lh2['vendor'].value_counts()
lh2
```

```python
vendor_per = pd.concat([lh_leg_cnt,lh2], axis='columns')
vendor_per
```

```python
vendor_per = vendor_per.fillna(0) 

vendor_per
```

```python
vendor_per.rename(columns={'fleet_vendor1':'leg_count','vendor':'breach_count'},inplace=True)
vendor_per
```

```python
vendor_per['on_time_delivered'] = (vendor_per['leg_count']) - (vendor_per['breach_count'])
vendor_per
```

```python
vendor_per['breach_count'] = vendor_per['breach_count'].astype('int64')
vendor_per['on_time_delivered'] = vendor_per['on_time_delivered'].astype('int64')
vendor_per
```

```python
vendor_per['performance %'] = (vendor_per['on_time_delivered']) / (vendor_per['leg_count']) * 100
vendor_per
```

```python
vendor_per.sort_values(['leg_count','performance %'], ascending=False)
```
